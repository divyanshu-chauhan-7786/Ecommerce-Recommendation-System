```python
import re


def extract_asin(url):
    """
    Extract ASIN from Amazon product URL.
    ASIN = 10-character alphanumeric code.
    """
    if pd.isna(url):
        return None

    
    match = re.search(r'/([A-Z0-9]{10})(?:[/?]|$)', str(url))
    if match:
        return match.group(1)


    match = re.search(r'dp/([A-Z0-9]{10})', str(url))
    if match:
        return match.group(1)

    return None



products["asin"] = products["link"].apply(extract_asin)
products[["name", "link", "asin"]].head(10)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def advanced_clean_text(text):
    if not isinstance(text, str):
        return ""

    text = text.lower()

    # Remove HTML tags
    text = re.sub(r'<.*?>', ' ', text)

    # Remove URLs 
    text = re.sub(r'http\S+|www\S+|https\S+', ' ', text)

    # Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9 ]', ' ', text)

    # Remove numbers if not needed
    text = re.sub(r'\d+', ' ', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    words = text.split()

    # Remove stopwords + lemmatize
    cleaned_words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words and len(word) > 2
    ]

    
    return " ".join(cleaned_words)
```

```python
products["clean_name"] = products["name"].apply(advanced_clean_text)
```

```python
products["combined_text"] = (
    products["clean_name"] + " " +
    products["main_category"].astype(str).str.lower() + " " +
    products["sub_category"].astype(str).str.lower()
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=4000,
    ngram_range=(1, 2)        
)

tfidf_matrix = tfidf.fit_transform(products["combined_text"])
```

```python

tfidf_dense = tfidf_matrix.astype("float32").toarray()
tfidf_dense.shape
```

```python
import faiss

dimension = tfidf_dense.shape[1]   
index = faiss.IndexFlatIP(dimension)  
```

```python
index.add(tfidf_dense)
```

```python
def search_product(query, top_n=5):
    query = query.lower().strip()
    results = products[products["clean_name"].str.contains(query)]
    return results.head(top_n)
```

```python
search_product(" kiroh")
```

```python
def recommend(product_index, top_n=10):
    # Get the vector of the selected product
    query_vec = tfidf_dense[product_index:product_index+1]

    # FAISS search
    scores, ids = index.search(query_vec, top_n)

    # Return similar products
    return products.iloc[ids[0]][
        ["name", "main_category", "sub_category", "discount_price", "link"]
    ]
```
